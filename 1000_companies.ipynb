```python
# Import libraries
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv('/content/1000_Companies.csv')
df.columns = df.columns.str.strip()  # remove extra spaces

# One-hot encode the 'State' column
df = pd.get_dummies(df, columns=['State'], drop_first=True)

# Define features and target
X = df.drop('Profit', axis=1)  # features: all except Profit
y = df['Profit']               # target: Profit

# Train Multiple Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Prepare new candidate data for prediction
# Candidate: 91694.48 R&D, 515841.3 Admin, 11931.24 Marketing, State=Florida
# Make sure columns match X columns (one-hot encoding)
import numpy as np

# Create zero array for all columns
new_data = np.zeros(X.shape[1])

# Set numeric features
new_data[X.columns.get_loc('R&D Spend')] = 91694.48
new_data[X.columns.get_loc('Administration')] = 515841.3
new_data[X.columns.get_loc('Marketing Spend')] = 11931.24

# Set the one-hot column for Florida State if exists
if 'State_Florida' in X.columns:
    new_data[X.columns.get_loc('State_Florida')] = 1

# Predict profit
predicted_profit = model.predict([new_data])
print("Predicted Profit for the company:", predicted_profit[0])
```
