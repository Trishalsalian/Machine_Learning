Predict canada's per capita income in year 2020. Use the data file canada_per_capita_income.csv file. If required, apply the necessary data processing steps. Using this build a regression model and predict the per capita income for canadian citizens in year 2020

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('/content/canada_per_capita_income.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
plt.scatter(df['year'], df['per capita income (US$)'])
plt.xlabel('Year')
plt.ylabel('Per Capita Income')
plt.show()
```

```python
X = df[['year']]   # independent variable
y = df['per capita income (US$)']   # dependent variable
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
prediction = model.predict([[2020]])
print("Predicted Per Capita Income for 2020:", prediction[0])
```

```python
print("Intercept:", model.intercept_)
print("Slope:", model.coef_[0])
```
